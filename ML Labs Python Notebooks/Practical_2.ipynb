**1. The if statement**

An `if` statement is a fundamental control flow statement in Python that allows you to execute a block of code only if a certain condition is true. It evaluates a boolean expression, and if the expression is true, the code inside the `if` block is executed. Otherwise, it is skipped.

```python
# if statement example

food = 'curry'

if (food == 'curry'):
    print("I love curry")

juice = "orange"

if (juice == 'orange'):
    print("I love orange")


if (food != 'curry' and juice != 'orange'):
    print("Not my favorite") # nothing will print because conditon return false
```

**2. The if else statement**

The `if-else` statement allows you to execute one block of code if a condition is true, and a different block of code if the condition is false. It provides a way to handle two possible outcomes based on a single condition.

```python
# if-else statement example

temperature = 25

if temperature > 30:
    print("It's a hot day!")
else:
    print("It's not too hot today.")

is_raining = False

if is_raining:
    print("Don't forget your umbrella!")
else:
    print("Enjoy the clear weather!")
```

**3. Chained Conditionals**

Chained conditionals, often implemented using `if`, `elif` (short for 'else if'), and `else` statements, allow you to check multiple conditions sequentially. Python evaluates conditions from top to bottom. The first condition that evaluates to `True` will have its corresponding code block executed, and the rest of the chain will be skipped. If no `if` or `elif` conditions are true, the `else` block (if present) will be executed.

```python
# Chained conditionals example
weather = "snowy"

if weather == "sunny":
    print("Let's go for a walk!")
elif weather == "rainy":
    print("Bring an umbrella and wear a raincoat.")
elif weather == "cloudy":
    print("It might be a good day for indoor activities.")
else:
    print("Stay warm and safe indoors!")

# checking grade
score = 75

if score >= 90:
    grade = "A"
elif score >= 80 and score < 90:
    grade = "B"
elif score >= 70 and score < 80:
    grade = "C"
elif score >= 60 and score < 70:
    grade = "D"
else:
    grade = "F"

print(f"With a score of {score}, your grade is: {grade}")
```

**4. Nested conditionals**

Nested conditionals occur when one `if` (or `elif`/`else`) statement is placed inside another `if` (or `elif`/`else`) statement. This allows for more complex decision-making, where a condition is checked only after an outer condition has been met. Indentation is crucial in Python to define which `if`/`elif`/`else` blocks belong to which outer statement.

```python
# Nested conditionals example
user_role = "admin"
is_logged_in = True

if is_logged_in:
    print("User is logged in.")

    if user_role == "admin":
        print("Welcome, administrator! You have full access.")
    elif user_role == "editor":
        print("Welcome, editor! You can create and modify content.")
    else:
        print("Welcome, user! You have basic access.")

else:
    print("Please log in to access the system.")


# With multiple nested conditions
weather = "sunny"
temperature = 28

if weather == "sunny":
    print("It's a sunny day!")

    if temperature > 25:
        print("It's hot, stay hydrated.")
    else:
        print("It's a pleasant day, enjoy!")

else:
    print("Consider checking the weather forecast.")
```

**5. Iteration**

Iteration is a fundamental concept in programming that involves repeatedly executing a block of code. In Python, this is primarily achieved using `for` loops and `while` loops. Iteration allows you to process sequences of data (like lists, strings, or tuples) or to execute a block of code a specific number of times or until a certain condition is met.

```python
# Iterating over a list
fruits = ["apple", "banana", "cherry"]
print("Iterating over fruits:")
for fruit in fruits:
    print(fruit)

# Iterating over a string
word = "Python"
print("\nIterating over a word:")
for char in word:
    print(char)

# Iterating using range()
print("\nCounting from 0 to 4:")
for i in range(5):
    print(i)

# Iteration using a while loop

count = 0
print("\nCounting with a while loop:")
while count < 3:
    print(f"Count: {count}")
    count += 1
```

**6. Re-assignment**

Re-assignment in programming refers to the act of changing the value associated with an existing variable. When a variable is re-assigned, its previous value is overwritten with the new value. This is a fundamental concept for dynamic data manipulation in Python and many other programming languages.

```python
# Variable re-assignment

x = 10
print(f"Initial value of x: {x}")

x = 20  # Re-assigning a new integer value to x
print(f"New value of x: {x}")

message = "Hello"
print(f"Initial message: {message}")

message = "Hello, Python!"  # Re-assigning a new string value
print(f"New message: {message}")

is_active = True
print(f"Initial status: {is_active}")

is_active = False  # Re-assigning a new boolean value
print(f"New status: {is_active}")

# Re-assigning a variable with a different data type
y = 5
print(f"Initial y (integer): {y}")

y = "five" # Re-assigning with a string
print(f"New y (string): {y}")
```

**7. The for loop**

The `for` loop in Python is used for iterating over a sequence (that is, a list, a tuple, a dictionary, a set, or a string). It executes a block of code for each item in the sequence. It's often used when you know in advance how many times you want to iterate or when you need to process all elements in a collection.

```python
# Loop through a list
colors = ["red", "green", "blue"]
print("Looping through colors:")
for color in colors:
    print(color)

# Loop through a string
word = "hello"
print("\nLooping through a word:")
for char in word:
    print(char)

# Loop using range()
print("\nCounting from 0 to 2:")
for i in range(3):
    print(i)

# Loop with index and value using enumerate
print("\nLooping with index and value:")
for index, fruit in enumerate(["apple", "banana"]):
    print(f"Index {index}: {fruit}")
```

**8. The while statement**

The `while` loop in Python repeatedly executes a block of code as long as a given condition is true. It is useful when you don't know the exact number of iterations beforehand, and the loop continues until the condition becomes false. It's important to ensure that the condition eventually becomes false to avoid an infinite loop.

```python
# Example of a while statement
count = 0
print("Counting up to 5:")
while count < 5:
    print(count)
    count += 1


# Another example: input validation
password = ""
while password != "secret":
    password = input("Enter the password: ")
    if password != "secret":
        print("Incorrect password. Try again.")
print("Access granted!")
```

**9. The break statement**

The `break` statement is used to terminate the loop entirely. When `break` is encountered inside a loop (either `for` or `while`), the loop is immediately exited, and the program control resumes at the statement immediately following the loop. It's often used to exit a loop prematurely when a certain condition is met.

```python
# Example of the break statement

print("Loop with break:")
for i in range(10):
    if i == 5:
        print(f"Breaking loop at i = {i}")
        break
    print(i)

# Break in a while loop
print("\nWhile loop with break:")
counter = 0
while True:
    print(counter)
    counter += 1
    if counter > 3:
        print(f"Breaking while loop at counter = {counter}")
        break
```

**10. The continue statement**

The `continue` statement is used to skip the rest of the current iteration of the loop and move to the next iteration. When `continue` is encountered, the code following it within the loop for the current iteration is skipped, and the loop proceeds with the next item or re-evaluates its condition. It's useful when you want to bypass certain parts of the loop's body for specific conditions without exiting the loop entirely.

```python
# Example of the continue statement

print("Loop with continue (skipping even numbers):")
for i in range(10):
    if i % 2 == 0:  # If i is an even number
        continue    # Skip the rest of this iteration
    print(i) # This will only print odd numbers

# Continue in a while loop
print("\nWhile loop with continue (skipping numbers divisible by 3):")
j = 0
while j < 10:
    j += 1
    if j % 3 == 0:
        continue # Skip printing if divisible by 3
    print(j)
```
