**1. Numpy**

NumPy is the fundamental package for scientific computing in Python. It provides a high-performance multidimensional array object and tools for working with these arrays.

```python
import numpy as np

# Creating 0D array
arr0 = np.array(42)
print("0D Array:", arr0)
print(type(arr0))

# Creating a 1D array
arr1 = np.array([1, 2, 3, 4, 5])
print("\n1D Array:", arr1)
print("Accessing element in 1D", arr1[2])

# Creating a 2D array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("\n2D Array:\n", arr2)
print("Accessing element in 2D", arr2[0][1])

# Basic Operations
print("\nSum of elements in arr1:", np.sum(arr1))
print("Mean of elements in arr1:", np.mean(arr1))
```

```python
# Accessing elements
print("First element:", arr1[0])
print("Slice of first 3 elements:", arr1[:3])

# Accessing 2D elements [row, column]
print("Element at row 1, col 2:", arr2[1, 2])
```

**Array Slicing**

Slices of arrays similar to Python lists but with more power for multidimensional data.

```python
# 1D Array Slicing
print("Elements from index 1 to 4:", arr1[1:4])
print("Elements from index 2 to end:", arr1[2:])
print("Last three elements:", arr1[-3:])

# 2D Array Slicing
# Syntax: arr[row_slice, column_slice]
print("\nFirst row, first two elements:", arr2[0, 0:2])
print("Second column from all rows:", arr2[:, 1])
print("Sub-matrix (first 2 rows, first 2 cols):\n", arr2[0:2, 0:2])
```

### **2. Pandas**

Pandas is a powerful library for data manipulation. The primary data structures in Pandas are Series (1D) and DataFrame (2D).

```python
import pandas as pd

# Reading a CSV file from the sample data folder
csv_path = "/content/sample_data/california_housing_train.csv"
housing_df = pd.read_csv(csv_path)

print("First 5 rows of the California Housing dataset:")
display(housing_df.head()) # also can use tail() for last 5 rows
```

```python
# Accessing specific columns from the housing DataFrame

# You can select a single column or multiple columns
rooms_column = housing_df['total_rooms']
print("First 5 entries of 'total_rooms' column:")
print(rooms_column.head())

# To select multiple columns, pass a list of column names
subset = housing_df[['latitude', 'longitude', 'median_house_value']]
print("\nSubset of specific columns:")
display(subset.head())
```

```python
import pandas as pd

# Creating a DataFrame from a dictionary
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "London", "Paris"]
}
df = pd.DataFrame(data)

print("DataFrame:\n", df)
print("\nAccessing the 'Age' column:\n", df['Age'])
print("\nAccessing the first row:\n", df.loc[0]) # to get first row
```

**Basic DataFrame Operations**

You can easily filter data, view statistics, and handle missing values.

```python
# Descriptive Statistics
print("Summary Statistics:\n", df.describe())

# Filtering data: People older than 28
older_than_28 = df[df['Age'] > 28]
print("\nPeople older than 28:\n", older_than_28)
```

### **3. Matplotlib**

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Creating a line plot
plt.plot(x, y, label='Linear Trend', color='blue', linestyle='--', marker='o')

# Adding titles and labels
plt.title("Simple Line Plot")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
plt.plot(y, linestyle = 'dotted')
plt.show()
```

**Bar Charts**

```python
categories = ['A', 'B', 'C', 'D']
values = [15, 24, 12, 30]

plt.bar(categories, values, color='teal')
plt.title("Categorical Comparison")
plt.show()
```

**Scatter Plots**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data
x = np.random.rand(10)
y = np.random.rand(10)

# Creating the scatter plot
plt.scatter(x, y, color='teal')
plt.title("Scatter Plot")
plt.xlabel("X variable")
plt.ylabel("Y variable")
plt.show()
```
