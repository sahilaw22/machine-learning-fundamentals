**2. String (or Text) Data Types**

**Note:** String can be created using single ( ' ), double ( " ) quotes and even triple ( """ or ' ' ' ) quotes.

```python
str = "hello python"
print(type(str))
```

```python

str1 = 'This is a string using single quotes.'
str2 = "This is a string using double quotes."

# Triple quotes can be used for multi-line strings
str3 = '''This is a
multi-line string
using triple single quotes.'''

# Triple double quotes can also be used for multi-line strings
str4 = """This is another
multi-line string
using triple double quotes."""

print(str1)
print(str2)
print("\n")
print(str3)
print("\n")
print(str4)
```

**Indexing and slicing in Strings**

```python
s = "hello" # Accessing single character using string

print(s[0])  #  h  e  l  l  o
print(s[1])  #  0  1  2  3  4
print(s[-1]) # -5 -4 -3 -2 -1
print(s[-5])
```

```python
slice = "Programming Language"

# slice[start : stop : step]
print(slice[0:])   # to print all
print("\n")
print(slice[::-1]) # to print in reverse order
print(slice[2:11:2]) # skip every 2nd character
print("\n")
print(slice[1:7])  # it start from index 1 and stop at 6 (exclude 7)
```

**3. Sequence Types**

Ordered collections. \
**• list** \
**• tuple**

**List** is an ordered collection of elements. It is mutable.

```python
# To make a list use square brackets []

list = [] # empty list

color = ["orange","red","green"]
odd_no = [1,3,5]

mixed = ["orange",2,"red",3,"green",4]

print(list)
print(color)
print(odd_no)
print(mixed)

# Accessing element using list

print(color[0])
print(color[1])
print(color[-1])
```

**Tuple** is also ordered  collection of elements but it is immutable.

```python
# Initializing an empty tuple
tup = ()

even = (2,4,6,8)
# even[0] = 12 is not allowed because of immutability
print(even)

name = ("rohan","kartik","angad")
print(name[0])
print(name[1])
print(name[2])
```

**4. Boolean (or bool) Data Type**

It represents one of two logical values: True (or 1) or False (or 0).

```python
# It is case sensitive mean True not equals to true
print(type(True))
print(type(False))

isTrue = True
isFalse = False

# 1 or any postive value means true
if (1 and isTrue):
   print("You are right")
# None 0 value means also true
if (not 0 and not isFalse):
       print("You are wrong")
```

**5. Set Types**

Set is a unordered collection of unique values.

```python

# It can removes duplicates automatically
# It can rearrange number in ascending order

s = {4,5,3,3,2,1}
s1 = set([4,5,5,3,2,1])
print("Both are valid in set\n",s,s1)

print("\n")
s2 = set(["hello","world","world"]) # using list
print(s2)
s3 = set("Hero") # using string
print(s3)

print("\n")
for i in s2:
   print(i,end=" ") #prints elements one by one

print("\n")
# check if exist in s2 return True or False
print("Geeks" in s2)
```

**6. Mapping (Dictionary or dict) Type**

The dictionary is the primary and only built-in mapping type in Python

```python
student = { 'name': 'Ali', "age":20 }

print(student["name"])
print(student.get("name"))

for i in student:
   print(i ,":", student[i]) # i display all keys and student[i] display values
```
